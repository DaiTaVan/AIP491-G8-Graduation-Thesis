# Get the current law table and law set

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# Initialize WebDriver (adjust path to your own WebDriver executable)
driver = webdriver.Chrome()

# Open the webpage
driver.get("https://luatvietnam.vn/tin-phap-luat/danh-muc-cac-bo-luat-luat-cua-viet-nam-230-18136-article.html?fbclid=IwY2xjawFG8ApleHRuA2FlbQIxMAABHdYX0Ggk_PuSmrypu3HdtEgrhUAlqp36uY1CTOiDKHaAXpbVmuO51gTBAQ_aem_B8M2vHf7AhX4scEMyV0ckg#:~:text=Hi%E1%BB%87n%2C%20Vi%E1%BB%87t%20Nam%20c%C3%B3%20266,B%E1%BB%99%20lu%E1%BA%ADt%20v%C3%A0%20Lu%E1%BA%ADt%20n%C3%A0y")

# Wait for the page to load (adjust the waiting time if needed)
driver.implicitly_wait(10)
```

```python
table = driver.find_element(By.TAG_NAME, "table")
```

```python
table
```

```python
# Extract table headers
headers = [header.text for header in table.find_elements(By.TAG_NAME, "th")]
headers
```

```python
rows = []
for row in table.find_elements(By.TAG_NAME, "tr"):
    cells = row.find_elements(By.TAG_NAME, "td")
    cell_content = []
    cell_content.append(cells[0].text)
    cell_content.append(cells[1].text)
    try:
        cell_content.append([cells[2].text, cells[2].find_element(By.TAG_NAME, "a").get_attribute('href')])
    except:
        cell_content.append(cells[2].text)
    if cells[3].text == '':
        cell_content.append('')
    else:
        try:
            cell_content.append([cells[3].text, cells[3].find_element(By.TAG_NAME, "a").get_attribute('href')])
        except:
            cell_content.append(cells[3].text)
    print(cell_content)
    rows.append(cell_content)
```

```python
formated_table = []
track_title = None
for i in range(len(rows)):
    if 'Bài viết liên quan' in rows[i]:
        track_title = rows[i]
    else:
        formated_table.append(dict(zip(track_title, rows[i])))
```

```python
formated_table
```

```python
import json
```

```python
json.dump(formated_table, open('bang_luat_bo_luat.json', 'w'), indent=2, ensure_ascii=False)
```
